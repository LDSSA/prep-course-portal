```python
import requests
import json


def submit(learning_unit: int, exercise_notebook: int, slackid: str, score: float) -> None:
    '''
    Submits the notebook.
    
    param url like "https://sub-nb-grades-collector.herokuapp.com/submit"
    param learning_unit like 0
    param exercise_notebook like 0
    param slackid like "UTS63FC02"
    param score like 16.0
    '''
    data = {
        "learning_unit": learning_unit,
        "exercise_notebook": exercise_notebook,
        "slackid": slackid,
        "score": score,
    }
    response = requests.post('http://127.0.0.1:8000/submissions/', json=data)
    print('Success' if response.ok else 'Whoopsie Daisy ' + response.text)
```
